```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
x1 = np.array([-2, -1, 0, 1, 2, 3, 4, 5, 6, 7])
y1 = np.array([9, 8, 7, 6, 5, 4, 3, 2, 1, 0])

x2 = np.array([-2, -1, 0, 1, 2, 3, 4, 5, 6, 7])
y2 = np.array([-3, -2, -1, 0, 1, 2, 3, 4, 5, 6])
```

```
print(plt.style.available)
```

```
plt.title("Graph:", size = 15)
plt.style.use('ggplot')
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.xlabel("Value of X", size = 10)
plt.ylabel("Value of y", size = 10)
plt.legend(["Equation-1", "Equation-2"], title = "Description", loc = 'upper right')
plt.show()
```
