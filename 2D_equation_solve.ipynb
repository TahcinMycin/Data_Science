```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
x1 = np.array([-2, 0, 2, 4, 6])
y1 = np.array([4, 2.5, 1, -0.5, -2])

x2 = np.array([-2, 0, 2, 4, 6])
y2 = np.array([-3, -1, 1, 3, 5])
```

```
x1, x2, y1, y2
```

```
plt.title("Graph:")
plt.style.use('classic')
plt.xlabel("X ---------------------------------->")
plt.ylabel("Y ---------------------------------->")
plt.plot(x1, y1, color = 'red')
plt.plot(x2, y2, color = 'green')
plt.legend(["Equation1", "Equation2"])
plt.show()
```
